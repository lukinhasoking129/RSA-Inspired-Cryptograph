```
from mpmath import mp
import random, math

# Precisão alta de pi
# Define a precisão para os cálculos de mpmath
mp.dps = 2000 # Aumentado para garantir dígitos suficientes
# Obtém o valor de pi com a precisão definida
piMil = mp.pi

def verificacao(num):
    """Verifica se o número não é divisível por 2, 3 ou 5."""
    if num % 2 == 0: return False
    if num % 3 == 0: return False
    if num % 5 == 0: return False
    return True

def getBlock():
    # Lógica original (gera um número de até 4 dígitos: produto de dois números de 2 dígitos)
    while True:
        # Gera um número aleatório para determinar a posição decimal do pi
        x = random.randint(1, 1800)
        num = 0
        sumVal = 1
        decimalPos = x
        for j in range(3):
            for i in range(2):
                digit = int((piMil * (10 ** decimalPos)) % 10)
                num = num * 10 + digit
                decimalPos += 1
            sumVal *= num
            num = 0

        # Verifica se o bloco gerado é seguro (dentro de um intervalo e não divisível por 2, 3 ou 5)
        if sumVal > 1000 and sumVal < 180000 and verificacao(sumVal):
            return sumVal

a = getBlock()
b = getBlock()
# Garante que o segundo bloco seja coprimo com o primeiro
while math.gcd(a, b) != 1:
    b = getBlock()

# Calcula N (produto dos blocos)
n = a * b

# Calcula o totiente composto
def totientesBlocks(blocks):
    phi = 1
    # Itera sobre cada bloco
    for block in blocks:
        n_temp = block
        result = n_temp
        p = 2
        # Calcula o totiente para cada bloco usando fatoração
        while p * p <= n_temp:
            if n_temp % p == 0:
                while n_temp % p == 0:
                    n_temp //= p
                result -= result // p
            p += 1
        if n_temp > 1:
            result -= result // n_temp
        # Multiplica os totientes dos blocos
        phi *= result
    return phi

# Calcula phi(N)
phiN = totientesBlocks([a, b])

# Escolher e e d válidos
e = 3
# Encontra um e coprimo com phiN
while math.gcd(e, phiN) != 1:
    e += 2

# Calcula d (inverso modular de e mod phiN)
d = pow(e, -1, phiN)

def criptografar(text, e, n):
    # Criptografa cada caractere do texto
    return [pow(ord(ch), e, n) for ch in text]


def descriptografar(mensagem, d, n):
    # Descriptografa cada valor numérico da mensagem
    return "".join(chr(pow(val, d, n)) for val in mensagem)

# Inicializa a mensagem criptografada como None
C = None
# Menu (inalterado)
menu = 0
while (menu != '3'):

  menu = input("\nMenu \n 1.Criptografar \n 2.Descriptografar \n 3.Sair \n\nEscolha Uma Opção\n")

  if menu == '1':

    if C == None:
      mensagem = input('Escreva Uma mensagem: ')
      print("Criptografando... Isso pode demorar alguns segundos")
      C = criptografar(mensagem,e,n)
      print(f"\nMensagem criptografada: {C} \nChave Publica:\nE={e}, \nChave Privada N={n} D={d}")
    else:
      print("Já existe Uma mensagem criptografada. Deseja substituí-la por uma nova?")
      opcao = input("S/N: ").upper()
      if opcao == "S":
        mensagem = input('Escreva Uma mensagem: ')
        print("Criptografando... Isso pode demorar alguns segundos")
        C = criptografar(mensagem,e,n)
        print(f"\nMensagem criptografada: {C} \nChave Publica:\nE={e}, \nChave Privada N={n} d={d}")
      else:
          print("Voltando ao Menu")
  elif menu == '2':
    if C is None:
        print("Nenhuma mensagem foi criptografada")
    else:
        M = eval(input("Digite a mensagem criptografada (ex: [123, 456, 789]): "))
        chaveN = int(input("Digite a Chave N: "))
        chaveE  = int(input("Digite a Chave E: "))
        chaveD = int(input("Digite a Chave D: "))
        if chaveN == n and chaveE == e and M == C and chaveD == d:
            print("Descriptografando...")
            M = descriptografar(M, d, n)
            print(f"\nMensagem Descriptografada: {M}")
            C = None
        else:
            print("Erro ao descriptografar!! , Verifique se as informações estão corretas")


  elif menu == '3':
    print("\nSaindo...")
    print("\nSessão encerrada")
  else:
    print("\nOpção Inválida")
```
